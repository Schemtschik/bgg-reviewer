```python
import pandas as pd
import re
```

```python
df = pd.read_csv('project_data-2.csv', sep='^')
df = df.fillna("['']")
decoder = lambda s: re.split("' ' |'\n ' ", s[2:-2])
df['mechanics'] = df['mechanics'].apply(decoder)
df['categories'] = df['categories'].apply(decoder)
df['game_designers'] = df['game_designers'].apply(decoder)
df['families'] = df['families'].apply(lambda x : x[2:-2].split("', '"))
df.head()
```

```python
df['id'] = df['game_id']
df['description'] = df['description_url']
df['designers'] = df['game_designers']
```

```python
games = df[['id', 'name', 'picture', 'description', 'player_min_age', 'players_min_amount', 'players_max_amount', 'duration', 'release_year', 'rating']]
games.to_csv('games.csv', index=None)
```

```python
types = [
#    ['mechanic', 'mechanics'],
#    ['category', 'categories'],
#    ['designer', 'designers'],
    ['family', 'families']
]
for t in types:
    cur_id = 0
    mapping = dict()
    out_df = pd.DataFrame(columns=['id', 'name'])
    mapping_df = pd.DataFrame(columns=['game_id', t[0] + '_id'])
    for index, row in df.iterrows():
        for entry in row[t[1]]:
            if entry not in mapping:
                mapping[entry] = cur_id
                out_df = out_df.append(pd.DataFrame({"id":[cur_id], "name":[entry]}))
                cur_id += 1        
            mapping_df = mapping_df.append(pd.DataFrame({"game_id":[row['id']], t[0] + '_id':[mapping[entry]]}))
    out_df.to_csv(t[1] + '.csv', index=None)
    mapping_df.to_csv('game_to_' + t[1] + '.csv', index=None)
    out_df
```
